3. For each of the following real-world problems, determine whether it is appropriate to utilise a supervised or unsupervised machine learning algorithm. Justify your answers:
* 3.1 Detecting anomalies in a manufacturing process using sensor data without prior knowledge of specific anomaly patterns. 
* 3.2 Predicting customer lifetime value based on historical transaction data and customer demographics. 
* 3.3 Segmenting customer demographics based on their purchase history, browsing behaviour, and preferences. 
* 3.4 Analysing social media posts to categorise them into different themes. 


3. Answer here
* 3.1 It is appropriate to use unsupervised machine learning algorithm as we do not have prior knowledge of specific anomaly patterns. Therefore, using a supervised algorithm using labelled data for training would not be possible. Also, anomalies represent deviations from normal behaviour so an unsupervised algorithm would be better to use here as the behaviour is not normal pattern and may not be understood well before.
* 3.2 It is appropriate to use supervised machine learning algorithm here as we need to predict customer lifetime value which is a numerical value but this is arrived using historical transaction data and customer demographics. These could be used as labelled inputs for learning and used to arrive at the output.
* 3.3 We have customer data of purchase history, browsing behaviour and preferences but however we do not have predefined labels within this data (for eg, highspenders, causal browsers). So, it would be better to use unsupervised machine learning algorithm to look at clusters within the existing data patterns and identify insights that were not initally present within the data to come up with segments of customer demographics.
* 3.4 We need to categorise social media posts into different themes. So, to do this we cannot use any supervised learning of which category a new social media post belongs to. It is better to use clustering through unsupervised machine learning algorithm to look at the data and uncover the hidden theme or topic so it can be categorised into appropriate category.

For each of the following real-world problems, determine whether it is appropriate to utilise semi-supervised machine learning algorithms. Justify your answers:
* 4.1 Predicting fraudulent financial transactions using a dataset where most transactions are labelled as fraudulent or legitimate.
* 4.2 Analysing customer satisfaction surveys where only a small portion of the data is labelled with satisfaction ratings.
* 4.3 Identifying spam emails in a dataset where the majority of emails are labelled.
* 4.4 Predicting the probability of default for credit card applicants based on their complete financial and credit-related information.


4. Answer here
* 4.1 Yes, it would be correct to utilise semi-supervised learning algorithm. The labelled data would have a transaction marked as either fradulent or genuine. So this data can be used as a training model. It provides foundation for algorithm to understand if a transaction is genuine or fradulent.But, in real world transactions there would be more of genuine transactions than fraudlent transactions which would be rare. In such scenario where there is scarcity of a labelled input semi supervised learning would help by looking at the data and capturing subtle patterns and unseen instances not shown by the labelled data alone. Since the data set is imbalanced with more genuine transactions and less fradulent transactions and a need to predict fradulent transactions it is better to use semi supervised learning algorithm.
* 4.2 Yes, it would be correct to utilise semi-supervised learning algorithm. As the labelled data is only a small portion of the data, even though limited would provide valuable information about customer satisfaction. It would serve as a foundation to predict satisfaction reatings. Semi-supervised learning would allow inclusion of unlabelled data in the training process. By using both labelled and unlabelled data it would be easier to capture nuances in the customer feedback.
* 4.3 It is appropriate to use a supervised algorithm here. The dataset contains mostly labelled emails.This provides substantial amount of data for learning whether an email is a spam or not. So this learning can be used based on existing patterns from labelled data to identify future emails and categorise them as a spam or not
* 4.4 It is appropriate to use a supervised algorithm here. The dataset would contain predefined information of the credit card applicants previous financial history and credit-card related information. Patterns for defaulting could be understood based on this data and this learning could be applied to the customer's prediction patterns of whether they would default or not.
